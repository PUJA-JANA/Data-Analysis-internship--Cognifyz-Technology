```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('Data analyst data.csv')
df.head()
```

## Task2: Votes Analysis

## Q1. Identify the restaurants with the highest and lowest number of votes.

```python
highest_voted_resturent= df.groupby('Restaurant Name')['Votes'].sum().sort_values()
highest_voted_resturent = highest_voted_resturent.idxmax()
print('The highest voted resturent are: ',highest_voted_resturent)
```

```python
lowest_voted_resturent= df.groupby('Restaurant Name')['Votes'].sum().sort_values()
lowest_voted_resturent = lowest_voted_resturent.idxmin()
print('The lowest voted resturent are: ',lowest_voted_resturent)
```

## Q2. Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
correlation = np.corrcoef(df['Votes'],df['Aggregate rating'])
correlation = np.round(correlation,2)
print(correlation)
```

```python
plt.figure(figsize= (10,5),dpi= 100)
sns.heatmap(correlation,annot= True, fmt= '.2f',linewidth= .4)
plt.show()
```

```python
sns.scatterplot(data= df,y= 'Votes',hue= 'Aggregate rating', x= 'Aggregate rating')
sns.regplot(data= df, x= 'Aggregate rating', y= 'Votes', scatter= False)
plt.title('Correlation between Aggregate Rating & Votes: 0.31')
plt.show()
print('As per result correlation value= 0.31, which indicates there is a moderate positive correlation between Aggregate rating and Votes')
```
