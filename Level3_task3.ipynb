```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('Data analyst data.csv')
df.head()
```

```python
df.columns
```

## Task3: Price Range vs. Online Delivery and Table Booking *

### Q1. Analyze if there is a relationship between the price range and the avail ability of online delivery and table booking?

```python
x1 = df['Has Online delivery']
x2 = df['Average Cost for two']
x3 = df['Has Table booking']
```

## Plot1 Online delivery with the Avg cost.
## Plot2 Table booking with the Avg cost

```python
plt.figure(figsize= (9,6))
plt.subplot(1,2,1)
sns.boxplot(x= x1, y= x2)
plt.xlabel('Online delivery')
plt.ylabel('Average Cost for two')
plt.title('Online Delivery affected by Cost')

plt.subplot(1,2,2)
sns.boxplot(x= x3, y= x2)
plt.xlabel('Table booking')
plt.ylabel('Average Cost for two')
plt.title('Table Booking affected by Cost')
plt.tight_layout()
plt.show()
print('Both the online delivery and table booking are nagetively impacted by the cost')
```

## Q2. Determine if higher-priced restaurants are more likely to offer these services.

### High price resturent which have avg cost are more then 10,000 did not provide online delivery as well as table booking facilities.

```python
# Lets check for the resturents have avg cost less then 10,000

plt.figure(figsize= (9,6))
plt.subplot(1,2,1)
sns.boxplot(x= x1,y= 'Average Cost for two',data= df.loc[df['Average Cost for two']< 10000])
plt.title('Online delivery by cost')

plt.subplot(1,2,2)
sns.boxplot(x= x3,y= 'Average Cost for two',data= df.loc[df['Average Cost for two']< 10000])
plt.title('Table booking by cost')
plt.tight_layout()
plt.show()
```

### The Online delivery facity present are those recturent where the avg cost less then 1000.
### The table booking facility is availabe in resturent with an avg cost rating from 1000 to 2000.
